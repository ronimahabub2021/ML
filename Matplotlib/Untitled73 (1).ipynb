```python
import matplotlib.pyplot as plt
```

Scatter Plots: Display individual data points on a graph to observe patterns or relationships.

```python
x=[1,2,3,4,5,6,7]
y=[8,9,10,11,12,13,14]
```

```python
plt.figure(figsize=(10,5))
plt.scatter(x,y,color='red',marker='h',label='scatter plot')
plt.plot(x,y,color='green',ls='--',label='Line Plot')
plt.legend(shadow=True,loc='best')
plt.grid(ls='--',alpha=0.5)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Scatter Plot')
plt.savefig('Test3')
plt.show()
```

```python
x=np.linspace(0,4*np.pi,100)
y=np.sin(x)
plt.plot(x,y)
plt.axhline(color='black',ls='--')
```

```python
x=np.linspace(0,4*np.pi,100)
y=np.cos(x)
plt.plot(x,y)
plt.axhline(color='black',ls='--')
```

```python
x=np.linspace(0,4*np.pi,100)
y1=np.sin(x)
y2=np.cos(x)
plt.plot(x,y1,label='Sin(x)',ls='-.',color='g')
plt.plot(x,y2,label='Cos(x)')
plt.legend()
plt.axhline(color='black',ls='--')
```

```python
x=np.arange(0,4*np.pi,0.1)
y1=np.sin(x)
y2=np.cos(x)
plt.plot(x,y1,label='Sin(x)',ls='-.',color='g')
plt.plot(x,y2,label='Cos(x)')
plt.legend()
plt.axhline(color='black',ls='--')
```
